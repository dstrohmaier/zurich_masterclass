# Do LLMs Produce Meaningful Output?

> "'When _I_ use a word,' Humpty Dumpty said in rather a scornful tone, 'it means just what I choose it to mean—neither more nor less.'" (Lewis Carroll, Through the Looking Glass)

## Questions to distinguish

- Do LLMs understand the meaning of text?
- Do LLMs understand the communicative intent behind a text?
- Do LLMs produce text responses indistiguishable from human responses?
    - Turing test
- Do LLMs produce text that carries meaning? (the current slides)


## To Carry Meaning

- Under what conditions does text carry meaning?
- Common assumption: Refereence to objects in the world has to be established
    - But see inferentialist counter-positions (Gubelmann 2024)

## Two Theories of What Establishes Reference


1. Internalism: states realise semantic properties that are causally effective,
    - In line with Titus (2024)
2. Externalism: the more popular option
    - Employed by Mandelkern & Linzen (2024) and Mollo & Millière (2025)

## Externalist Response

- Meaning ain't in the head
    - Putnam (1973): Twin earth experiment
    - Kripke (1980): Names as rigid designators
- World-Word connection: Reference
- External facts establish reference


## Requirements for Reference

- Causal history
    - Interaction with object (water vs. twater)
    - Baptism event and causal chain of transfer
    - Selection pressure 
- Social connection
    - Deference to experts (elm-beech example)
    - Participating in speech community (Burge 1972, used in Mandelkern & Linzen 2024)

## Any Internal Conditions?

- Narrow content (Titus 2024)
- Communicative intent (Bender & Koller 2020)

## Questions

- There is a causal chain, is it the right one?
- Can LLMs defer to others?
- Externalism covers names and natural kind terms well
    - What about verbs? Are the verbs produced by LLMs meaningful?
    - What about indexicals? What does the "here" produced by an LLM mean?

## What's the Point?

- What difference does it make whether the externalist conditions are met?
- No difference to behaviour, no difference to work in AI research, no obvious ethic implications...
