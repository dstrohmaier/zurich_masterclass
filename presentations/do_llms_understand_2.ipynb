# Do LLMs Understand Meaning?

> “Abduction is the process of forming explanatory hypotheses. It is the only logical operation which introduces any new idea” (C. S. Peirce)

- "Does ChatGPT Have Semantic Understanding? A Problem with the Statistics-of-Occurrence Strategy" by Titus (2024)

- Discussion of whether LLMs understand meaning (ChatGPT)

## Structure of the Paper

1. Presents functioning criterion
2. Behavioural evidence does not establish that functioning crtierion is met
3. Alternative: Statistical hypothesis (better explanation)
4. Considers counter-arguments

## Functioning Criterion

> "A system with semantic understanding functions in ways that are casually explainable by appeal to the semantic relationships among its states and processes with semantic content, and this functioning typically drives the evolution of these and and processes as well as the system's overt behavior." (Titus 2024, p. 3)

## Key Features

- Causal role of states with semantic content
- Shorter formulation: functional sensitivity to semantic properties
- Not (just) behavioural
- Do we need to attribute semantic sensitivity?

## The Alternative: Statistical Hypothesis

> "if it is plausible that merely attributing sensitivity to word co-occurrence statistics is sufficient to explin SOM's production of meaning semblant behavior, then such an account is preferable, because it is simpler and more parsimonious." (Titus 2024, p. 5)

## The Argument

- Abduction/Inference to the best explanation 
- Relies on underdetermination by behavioural data
    - At least by available behavioural data 

## Multi-Modal Models

- Discusses multi-modal models in section 5.2
- Should consider empirical investigation

> "But the system is actually unlikely to be tracking 90-degree angles of images per se." (Titus 2024, p. 11)

> "We find that each curve detector in the second layer of the ViT consistently activates for a specific angle." (Kim et al. 2025)


See also chapter 3 of Buckner 2024

## Comparison

|                          | Bender & Koller 2020           | Titus 2024                         |
|--------------------------|--------------------------------|------------------------------------|
| _Models considered_      | GPT-2, BERT                    | GPT-3, ChatGPT, Multimodal         |
| _Understanding requires_ | Learning form-meaning relation | Sensitivity to semantic properties |
| _Problem_                | Wrong data                     | Wrong internal processing          |
| _Argument_               | Deductive + Thought experiment | Inference to the best explanation  |
| _Strength of argument_   | Impossibility of understanding | Other explanations superior        |

# Philosophical Issues

- Does sensitivity to semantic properties entail narrow content?
- Paper somewhat unclear on behavioural evidence (page 5, a large amount of behavioural data appears relevant. Why?)
- Formulation of functioning criterion mentions evolution. Why?
- Why is it more parsimonious to explain the behavioural skills of models by attributing them the ability to track very complex statistical patterns rather than simple semantic properties?
