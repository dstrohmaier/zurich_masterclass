# The word2vec Moment

- Neural networks breakthrough in the early 2010s
- Images: AlexNet
- Text: word2vec (Mikolov, Chen et al. 2013)

## What Is word2vec?

- word2vec is a set of algorithms for learning word representations
- word2vec trains on context to create representations

> "You shall know a word by the company it keeps" (Firth 1957, p. 11)

- Representations are continuous
- Providing notion of similarity

### Properties of word2vec

- Neural network
- Extremly simple (not deep): 2 vectors for every word
- Useful, e.g. sentiment analysis of movie reviews

# word2vec and Linguistic Regularities

- word2vec was created for applications
- Language modelling as a task to train general representations of words
- Representations encoded linguistic regularities (Mikolov, Yih, Zweig 2013)

Regularities are encoded as parallelograms

## Semantic
- king:man::queen:woman
## Grammatical
- go:went::eat::ate

## Vector Offset Method

1. Take vector for base word: "king" $ \rightarrow [ 0.13,  0.03,  0.01\ldots ]$ 
2. Subtract negative vector: "man" $\rightarrow [ 0.33,  0.13,  0.03\ldots]$
3. Add positive vector: "woman" $\rightarrow [0.24, -0.08, -0.1\ldots]$
4. Find closest vector in space: "queen" $\rightarrow [0.01, -0.14, -0.07\ldots]$

## Remarks on the Offset Method

- Why? Trained on a seemingly unrelated task
- Closest vector (cosine similarity / angle)
- Directions are meaningful
- What does it tell us about how the model works?
  + Does an encoded bias matter? Can be misleading, see Nissim et al. (2020)
