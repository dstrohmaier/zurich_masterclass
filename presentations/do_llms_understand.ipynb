# Do LLMs Understand Meaning?

> "Wenn ein Löwe sprechen könnte, wir könnten ihn nicht verstehen" (Wittgenstein, Philosophische Untersuchungen II, xi)

- "Climbing towards NLU: On Meaning, Form, and Understanding in the Age of Data" by Bender & Koller (2020)
- Early discussion of whether LMs understand meaning (GPT-2, BERT)
- Famous for its octopus thought experiment

## Structure of the Paper

1. Proposes a "working definition" of _meaning_
2. Argue that meaning cannot be learned from form alone
3. Provide octopus thought experiment (and others)
4. Positioning within theory (Climbing the _right_ hill?)
5. React to some possible responses

## Definition of Meaning

- Gricean approach
- Occasion meaning as a relation: $ M \subseteq E \times I $
- Conventional linguistic system: $ C \subseteq E \times S $

<br>
<br>

| Symbol | Description                  |
|--------|------------------------------|
| $E$    | linguistic expressions       |
| $I$    | communicative intents        |
| $S$    | standing meanings (abstract) |

## Key Properties of this Critique

- Targets understanding
    - Can the octopus generate meaningful responses?
- Empiricist
- Not the right _data_

> "[...] if the training data is only form form, there is no sufficient signal to learn the relation $M$ between the that form and the non-linguistic intent of human language users, nor $C$ between form and the standing meaning the linguistic systems assigns to each form." (Bender & Koller 2020, p. 5187)
