## A Historical Sketch


- Starting point: Searle's (1980) Chinese room thought experiment
- Grounding problem introduced by Harnad (1990)
  - Chinese dictionary thought experiment &rarr; carousel of representations
- To get out of the carousel: _ground_ the represetations

## The AI Context of the 80s and 90s

- Symbolic, logic-based AI dominated AI research
- How do you connect DOG to dogs? 
- Neural models had recently re-emerged
    - 1986: Parallel Distributed Processing by Rumelhart, McClelland and the PDP research group
- Harnad suggested combined models (neuro-symbolic)

## The Original Grounding Problem

- How to connect symbols in reasoning systems to the real world
- How to give symbols _intrinsic_ meaning
- Build reasoning systems on top of vision, movement, etc. 

## The Original Solution by Harnad

Build a model that can fulfill 6 tasks


| # | Description                       |
|---|-----------------------------------|
| 1 | Discriminate between inputs       |
| 2 | Manipulate perceptual inputs      |
| 3 | Identify inputs / assign labels   |
| 4 | Describe inputs (internally)      |
| 5 | Produce descriptions (linguistic) |
| 6 | Respond to descriptions           |

## Two Paths

1. Easy grounding problem: Develop technology that goes beyond the carousel of representations, engaging and interacting competently with the world
2. Hard grounding problem: Establish and implement conditions for intrinsic meaning

(similar easy vs. hard terminology in Müller 2011)

## The Easy Grounding Problem in the Age of LLMs

- Commonly investigated in computer science/AI research
    - Multi-modality
    - Factual grounding (linking to databases etc.)
    - Robotics 
- Easy in the sense that the easy problem of consciousness is easy (not really)

## The Hard Grounding Problem in Age of LLMs

- "The Vector Grounding Problem" by Mollo & Millière (2025)
- Grounding of vectors. Do the representations carry meaning?
- Are vectors even plausible carriers of lexical meaning?

## The Externalist Approach

- Mollo & Millière follow externalist (teleosemantic?) mainstream
- Externalist criteria: Causal history + selection pressure
    - Which selection?
- RLHF creates selection pressure 

## The Swamp-LLM Thought Experiments

in Mollo & Millière (2025)

- Assume you have trained, grounded LLM
- Assume you randomly initialise an LLM with the same weights
- The latter would not be causally grounded

## Discussion Questions

- If we had a look-up table with the right causal history and selection pressure, would it produce meaningful text?
- Is there a difference between a swamp computer and a swamp LLM?

## Discussion Questions II

- How could we ever decide between two subtly different accounts of externalism for LLMs?
    - How do we determine the degree of selection pressure required?
    - Are our intuitions fine-grained enough for that?

## Why Care about the Hard Grounding Problem?

- What differences does it make if vectors are grounded?
- Mollo & Millière (2025): None for performance!

> "The quality of outputs and their status as intrinsically meaningful are orthogonal
  issues." (Mollo & Millière 2025)

- Externalism also decouples meaning and understanding

## Why Care?

- Makes no difference for CS benchmarks
- Makes no difference to users
- Makes no difference for AI ethics (as far as I can tell)

## Does it Make a Difference for Philosphers?

- Does it have implications for philosophy of language whether LLMs produced meaningful text?
- Is carrying intrinsic meaning epiphenomenal?
- Can one understand a text that has no meaning?
