## Goals

1. Understand architectural principles behind language models
2. Provide intuitions for how representations might capture meaning
3. Connect the computer science and philosophical debates about the meaningfulness of content
4. Develop the philosophical impact of cutting edge issues

## Connections to  topics in philosophy

- Philosophy of language: What are the conditions for words to carry meaning?
- Philosophy of science: What counts as a successful model of meaning? How to establish its correctness?
- Philosophy of cognitive science: What notion of representation is at play? 
- AI Ethics: How to interpret patterns in AI models (bias) and their limits?
